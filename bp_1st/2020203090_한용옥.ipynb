```python
cake_shape = "*"
space_shape = " "
```

```python
def make_column(num, max_width):
    space = (max_width - 2*num + 1)//2
    return (space_shape*space) + (cake_shape + space_shape)*num + (space_shape*(space - 1))
```

```python
def make_floor(num, max_width):
    temp_width = 2*(num + 2) + 1
    temp_space_width = (max_width - temp_width)//2
    return space_shape*temp_space_width + cake_shape*temp_width + space_shape*temp_space_width
```

```python
def make_cake(num):
    max_width = 2*(num + 3) + 1
    for i in range(1, num + 1):
        column = make_column(i, max_width)
        print(column + '\n' + column)
        if i != num:
            print(make_floor(i, max_width))
        else:
            print(cake_shape*max_width)
```

```python
def receive_input():
    num = int(input("N단 케이크의 N 입력 (1 이상 10 이하의 N 입력) : "))
    while not 1 <= num <= 10:
        num = int(input("잘못된 입력입니다. 다시 입력하세요\nN단 케이크의 N 입력 (1 이상 10 이하의 N 입력) : "))
    return num
```

```python
def main():
    while True:
        a = int(input("도형 그리기\n    1. 도형 그리기\n    2. 종료\n숫자를 입력해 선택하세요 :  "))
        if a == 2:
            break
        elif a == 1:
            make_cake(receive_input())
        else:
            print("잘못된 입력입니다. 다시 입력하세요. ")
```

```python
main()
```
